# Which Factor Has a Greater Impact on Net Migration?

##1. Project Introduction

###1.1 Introduction

Population migration has always been an important topic in urban research. Population migration brings rich labor resources and human resources to cities, and population outflow may lead to urban hollowing out, decline of economic vitality and other problems. Therefore, accurately understanding the influencing factors and making accurate predictions of population migration is a worthy question.

This study, selected the United States, analyzed the data from 50 US states between 2010 and 2020, emphasize the significance of accurately modeling and predicting migration patterns using a combination of census, economic, and industry data. The integration of machine learning in this process could potentially enhance the accuracy and reliability of migration forecasts, making this research question highly valuable for understanding demographic changes and informing policy decisions

### 1.2 Research Question

Which Factor Has a Greater Impact on Net Migration?

###1.3 Literature Review


Research in migration commonly considers migration as a portfolio of activities that can be specialized or intensified in particular activities, diversified across different activities, or include migration (short or long-term) to other locations (Barrett et al., 2001）

Starting with the initial exploration of the push-pull theory, the demographic theory has explained migration and its activities in the context of livelihood by analyzing the "push" and "attraction" factors, and how the "intervention" factors limit any migration behavior (Bogue 1969). Subsequently, the theory modified the intervention factors by introducing the concept of "anchor point" (Moon 1995), forming what is now called the "push-pull-anchor" (PPM) migration theory. Anchors can be widely understood as specific aspects of an individual's "place utility" linking them to somewhere, including family or assets, love for the place, or other factors causing inertia or status quo bias (Adams & Adger, 2013).


Studies over the past decade have included migration (including commuting and temporary or seasonal migration) as part of a legitimate livelihood strategy, rather than to be avoided and prevented, provided that voluntary movement to improved improvements is more popular than involuntary forced movements (Adger et al., 2009). At the same time, advances in computing power have led to a shift in the way in which migration is simulated on computers, from focusing solely on location to focusing directly on population. This approach shift not only deepens our understanding of migration dynamics, but also provides researchers with finer tools to analyze and predict migration patterns

##5.1 Set Training Set and Testing Set

Use sklearn.model_selection module to split a dataset into training and testing sets, set random seeds to a fixed value to ensure a stable output.

```python
random_state_split = 50
x=mdf[[ 'BIRTHS', 'DEATHS','POPESTIMATE',
       'Total exemptions', 'Median AGI', 'Mean AGI', 'AveragePrice','Estimate!!Median selected monthly owner costs (dollars)!!Total',
       'Estimate!!Median selected monthly owner costs (dollars)!!Housing units with a mortgage (dollars)',
       'Estimate!!Median selected monthly owner costs (dollars)!!Housing units without a mortgage (dollars)']]
train_x, test_x, train_y, test_y = train_test_split(
    mdf[[ 'BIRTHS', 'DEATHS','POPESTIMATE',
       'Total exemptions', 'Median AGI', 'Mean AGI', 'AveragePrice','Estimate!!Median selected monthly owner costs (dollars)!!Total',
       'Estimate!!Median selected monthly owner costs (dollars)!!Housing units with a mortgage (dollars)',
       'Estimate!!Median selected monthly owner costs (dollars)!!Housing units without a mortgage (dollars)']],
    mdf['NETMIG'],
    random_state=random_state_split
)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)

# check the index of train_x and train_y - they should be identical. The index indicates which rows from the original data.

print(train_x.index.identical(train_y.index))
print(test_x.index.identical(test_y.index))
```

###5.2 CART

**CART** is a decision tree algorithm that can be used for both classification problems (classification trees) and regression problems (regression trees). The principle involves recursively splitting the data set into smaller subsets while at the same time developing a decision tree. Each node in the tree represents a question or condition that splits the data into two subsets, hence the decision tree's binary splitting at each step.Similarly, **Random Forest** builds many decision trees and merges them together to get a more accurate and stable prediction. Here, we use **GridSearchCV** to optimize the hyperparameters

```python
# a CART using default settings
cart_default = DecisionTreeRegressor(random_state=0)
cart_default.fit(train_x, train_y)
# print the tree depth
print("Tree depth: {}".format(cart_default.get_depth()))
```

```python
# values of max_depth and min_samples_split
hyperparameters = {'max_depth':[5,10,15,20,25], 'min_samples_split':[2,4,6,8,10]}

randomState_dt = 10000
dt = DecisionTreeRegressor(random_state=randomState_dt)

# cv=5 by default, which means 5-fold cross-validation
clf = GridSearchCV(dt, hyperparameters)

clf.fit(train_x, train_y)

# we can query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (clf.best_params_)
print ("The best score is: ")
print (clf.best_score_)
```

```python
dt_final = DecisionTreeRegressor(max_depth=clf.best_params_['max_depth'], min_samples_split=clf.best_params_['min_samples_split'], random_state=randomState_dt)
dt_final.fit(train_x, train_y)
```

```python
print("R2 on the training data:")
print(dt_final.score(X=train_x, y=train_y))
print("R2 on the testing data:")
print(dt_final.score(X=test_x, y=test_y))
print("RMSE on the training data:")
print(mean_squared_error(train_y, dt_final.predict(train_x), squared=False))
print("RMSE on the testing data:")
print(mean_squared_error(test_y, dt_final.predict(test_x), squared=False))
```

###5.3 Random Forest Regression

```python
# values of max_depth and min_samples_split
hyperparameters = {'max_depth':[5,10,20,30,40], 'min_samples_split':[2,4,6,8,10]}

randomState_dt = 10000
rf = RandomForestRegressor(random_state=randomState_dt)

# cv=5 by default, which means 5-fold cross-validation
clf = GridSearchCV(rf, hyperparameters)

clf.fit(train_x, train_y)

# we can query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (clf.best_params_)
print ("The best score is: ")
print (clf.best_score_)
```

```python
rf_final = RandomForestRegressor(max_depth=clf.best_params_['max_depth'], min_samples_split=clf.best_params_['min_samples_split'], random_state=randomState_dt)
rf_final.fit(train_x, train_y)
```

```python
print("R2 on the training data:")
print(rf_final.score(X=train_x, y=train_y))
print("R2 on the testing data:")
print(rf_final.score(X=test_x, y=test_y))
print("RMSE on the training data:")
print(mean_squared_error(train_y, rf_final.predict(train_x), squared=False))
print("RMSE on the testing data:")
print(mean_squared_error(test_y, rf_final.predict(test_x), squared=False))
```

###5.4 XGboost

**XGBoost** Is a promotion-based ensemble learning method that improves prediction by gradually building models. Each new model focuses on the underpredicted portion of the previous model

```python
# values of max_depth and min_samples_split
hyperparameters = {'max_depth':[5,10,15,20,30], 'n_estimators':[20,50,100,150,200]}

randomState_xgb = 125
xgb = XGBRegressor(random_state=randomState_xgb)

# cv=5 by default, which means 5-fold cross-validation
gscv_xgb = GridSearchCV(xgb, hyperparameters)

gscv_xgb.fit(train_x, train_y)

# we can query the best parameter value and its accuracy score
print ("The best parameter value is: ")
print (gscv_xgb.best_params_)
print ("The best score is: ")
print (gscv_xgb.best_score_)
```

```python
xgb_final = XGBRegressor(max_depth=gscv_xgb.best_params_['max_depth'], n_estimators=gscv_xgb.best_params_['n_estimators'], random_state=randomState_xgb)
xgb_final.fit(train_x, train_y)
```

```python
print("R2 on the training data:")
print(xgb_final.score(X=train_x, y=train_y))
print("R2 on the testing data:")
print(xgb_final.score(X=test_x, y=test_y))
print("RMSE on the training data:")
print(mean_squared_error(train_y, xgb_final.predict(train_x), squared=False))
print("RMSE on the testing data:")
print(mean_squared_error(test_y, xgb_final.predict(test_x), squared=False))
```

###5.5 Lasso

**LASSO** performs variable selection by imposing an L1 penalty on the coefficients. This penalty can reduce the number of variables in the model, and thus reduce the problem of multicollinearity

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

# 设定超参数的范围
hyperparameters = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]}

# 创建Lasso回归模型实例
lasso = Lasso(random_state=random_state_split)

# 创建GridSearchCV实例来寻找最佳参数
clf = GridSearchCV(lasso, hyperparameters, cv=5)

# 在训练集上训练模型
clf.fit(train_x, train_y)

# 获取最佳参数
best_params = clf.best_params_
print('Best parameters found: ', best_params)

# 使用最佳参数创建新的Lasso实例
best_lasso = Lasso(alpha=best_params['alpha'], random_state=random_state_split)

# 在训练集上训练模型
best_lasso.fit(train_x, train_y)
```

```python
print("R2 on the training data:")
print(best_lasso.score(X=train_x, y=train_y))
print("R2 on the testing data:")
print(best_lasso.score(X=test_x, y=test_y))
print("RMSE on the training data:")
print(mean_squared_error(train_y, best_lasso.predict(train_x), squared=False))
print("RMSE on the testing data:")
print(mean_squared_error(test_y, best_lasso.predict(test_x), squared=False))
```
